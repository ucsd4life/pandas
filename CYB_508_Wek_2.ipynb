```python
tcp_mode = df["tcp_packets"].mode()[0]

df.tcp_packets.fillna(tcp_mode, inplace = True)

dist_port_mode = df["dist_port_tcp"].mode()[0]

df.dist_port_tcp.fillna(dist_port_mode, inplace = True)

external_ips_mode = df["external_ips"].mode()[0]

df.external_ips.fillna(external_ips_mode, inplace = True)

vulume_bytes_mode = df["vulume_bytes"].mode()[0]

df.vulume_bytes.fillna(vulume_bytes_mode, inplace = True)

udp_mean = df["udp_packets"].mean()

df.udp_packets.fillna(udp_mean, inplace = True)

tcp_urg_mode = df.["tcp_urg_packet"].mode()[0]

df.isnull().sum()
```
